### フォルダの生成

````python
import os

parent_name = "Datasets"
base_name = "Data"
num_dataset = 100

for i in range(num_dataset):
    path = os.path.join(parent_name, "{0}_{1:03d}".format(base_name, i+1))
    if os.path.exists(path):
        pass
    else:
        os.mkdir(path)
````

作業用フォルダに100個フォルダが生成されたことを確認してください。

### データの生成と保存
本当はデータ生成と同時にすりゃいいけど、見やすさのために分けてます。

````python
import numpy as np
import pandas as pd

mu = (0,0)
cov = np.diag([1,1])
np.random.seed(181003)

for i in range(num_dataset):
    cov[0,1] = i/num_dataset
    cov[1,0] = i/num_dataset
    dataset = np.random.multivariate_normal(mu, cov, size=1000)
    df = pd.DataFrame(dataset, columns=["x", "y"])
    
    path = os.path.join(parent_name, "{0}_{1:03d}".format(base_name, i+1))
    df.to_csv(os.path.join(path, "data_{}.csv".format(i+1)))
````

各フォルダにデータが生成されたことを確認してください。  
これにて準備完了です。

# フォルダ＆ファイル検索

````python
import os,glob
from collections import OrderedDict

parent_path = "Datasets"

dir_list = glob.glob1(parent_path, "*")
print(dir_list)
````

詳細は「正規表現」を検索（使い方は難解なので、正規表現とは何かくらいわかってればいい）。

"Data_"の文字列を含んだものだけ見つけたい

````python
parent_path = "Datasets"

dir_list = glob.glob1(parent_path, "*Data_*")
print(dir_list)
````

「Data\_」という文字列を持っていて、「Data\_」の前後の文字は何でもよいという意味

適当にフォルダにゴミファイルを投げ込んで上記の比較をすると違いが判る。

__さらに下層にあるデータファイル(csvファイル)を検索、読み込み__

````python
parent_path = "Datasets"
dir_list = glob.glob1(parent_path, "*Data_*")
df_dict = OrderedDict()   # 順序をキープしてくれるディクショナリ

for dir_name in dir_list:
    path = os.path.join(parent_path, dir_name)
    fname = glob.glob1(path, "data_*.csv")[0]   # globは要素が1個でもリストで返してくるので[0]でスライス
    fpath = os.path.join(path, fname)
    df_dict[dir_name] = pd.read_csv(fpath, header=0, index_col=0)   # header=Noneにすれば1行目もデータとして読む
````

## データを一気に読み込んだ恩恵

そりゃあ一気に解析できることでしょう。  
生成したデータは2次元正規分布に従ってデータが生成されています。  
各データセットについて平均と分散を計算してみましょう。

````python
aves = pnl.mean(axis=1).T   # 0, 1, 2の順にitem, major, minor, Tは行列の転置
aves
````

````python
vars =  pnl.var(axis=1).T   # 不偏分散を教えるためにわざとddof=1なんてやってたけど、実はデフォルトでddof=1なので書かなくても良い
vars
````

じゃあそれぞれ、元のフォルダに結果を保存してきましょう。  
この結果をバラさずにまとめフォルダ的なところに保存したほうが便利だけど、まぁ今回は練習ってことで面倒くさいことをしています。  
マスターすれば、同じ操作を複数データセットに一気にできるようになるでしょう。

column名にフォルダ名をそのままつかったので、保存先の指定もカラム名を使えば簡単にできますね（順番が狂っていても困らないということ）。

````python
indecies = list(aves.index)

for dir_name in indecies:
    path = os.path.join(parent_path, dir_name)
    data_ave = aves.loc[dir_name,:]
    data_var = vars.loc[dir_name,:]
    
    data_ave.to_csv(os.path.join(path, "ave.csv"))
    data_var.to_csv(os.path.join(path, "var.csv"))
````

各フォルダにデータが保存されたのを確認しましょう。数秒で100個のフォルダに入っているデータを集計し、各フォルダに計算結果を格納できました。
