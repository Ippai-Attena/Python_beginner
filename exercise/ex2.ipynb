# 線形単回帰
取ってきたデータから検量線を作ります。  
これから使うscikit-learnは、すべてnumpyのmatrix型で値を受け取ることを前提としています。  
Pandasデータフレームのままだとエラーになるので、numpyに戻してから渡してやりましょう。  
＊array型ではなことに注意。  
このサイトがとても親切。  
https://pythondatascience.plavox.info/scikit-learn/%E7%B7%9A%E5%BD%A2%E5%9B%9E%E5%B8%B0

````python
np.matrix(calib1.loc[:,x_name].values).T
````

````python
from sklearn.linear_model import LinearRegression

model1 = LinearRegression()   # モデルを設定とともに実体化（インスタンス化）する。デフォルトでよければLinearRegression()で良い
model2 = LinearRegression()

X1 = np.matrix(calib1.loc[:,x_name].values).T    # Tは転置のメソッド。ベクトルは縦が標準なので、縦ベクトルにしている。
Y1 = np.matrix(calib1.loc[:,y_name].values).T
X2 = np.matrix(calib2.loc[:,x_name].values).T
Y2 = np.matrix(calib2.loc[:,y_name].values).T

model1.fit(X1, Y1)   # 学習
model2.fit(X2, Y2)   
````

fitで線形回帰モデルの学習をしてくれます。  
ただの1次関数なので傾きを切片を計算しただけですけど。  
R2値とかも保持しているので見ていきましょう。

````python
# 回帰係数
print("slope")
print("calib1\t:", model1.coef_)
print("calib2\t:", model2.coef_)
 
# 切片 (誤差)
print("\nintercept")
print("calib1\t:", model1.intercept_)
print("calib2\t:", model2.intercept_)
 
# 決定係数
print("\nR2")
print("calib1\t:", model1.score(X1, Y1))
print("calib2\t:", model2.score(X2, Y2))
````

単回帰だけじゃなく重回帰もいけるので、係数は2次元配列、切片は1次元配列で返されています。  
扱いに注意しましょう。  

では得られた回帰直線をデータに重ね書きして、うまく検量線が引けたか確認しましょう。  
ついでに綺麗なグラフを作って出力しておけば、そのまま使いまわせますね。

````python
plt.rcParams["font.size"] = 16
plt.rcParams["font.family"] = "serif"
figuresize = (6,6)

xrange1 = np.matrix(np.linspace(0, 1000, 1000)).T
xrange2 = np.matrix(np.linspace(0, 200, 1000)).T

pred_y1 = model1.predict(xrange1)   # 予測はpredictメソッドでできる。
pred_y2 = model2.predict(xrange2)

# calib1のプロット
label_eq1 = "y = {}x{}".format(model1.coef_[0,0], model1.intercept_[0])
label_eq2 = "y = {}x+{}".format(model2.coef_[0,0], model2.intercept_[0])
print(label_eq1)
print(label_eq2)

plt.figure(figsize=figuresize)   # 新しいグラフ描画領域を作成
plt.plot(xrange1, pred_y1, color="k", lw=2, label=label_eq)
plt.scatter(x1, y1, s=100, edgecolors="k")
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()
ax.set_aspect((xlim[1]-xlim[0])/(ylim[1]-ylim[0]))
plt.title("Calibration curve")
plt.xlabel("Concentration ($\mu$g/mL)")
plt.ylabel("sample/I.S. (A.U.)")
plt.tight_layout()
plt.savefig(os.path.join(dir_name,"calib1.png"), dpi=300, transparent=True)   # transparent=Trueで、背景を透明にしてます。pngじゃないと透明にならないので注意。
plt.clf()
plt.close()


# calib2のプロット(ほぼ上のコピペ)
plt.figure(figsize=figuresize)   # 新しいグラフ描画領域を作成
plt.plot(xrange2, pred_y2, color="k", lw=2)
plt.scatter(x2, y2, s=100, edgecolors="k")
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()
ax.set_aspect((xlim[1]-xlim[0])/(ylim[1]-ylim[0]))
plt.title("Calibration curve")
plt.xlabel("Concentration ($\mu$g/mL)")
plt.ylabel("sample/I.S. (A.U.)")
plt.tight_layout()
plt.savefig(os.path.join(dir_name, "calib2.png"), dpi=300, transparent=True)
plt.clf()
plt.close()
````

# 求めた検量線を使って定量していく
今回の測定データはpeak2が全て内標なので、一気に面積比を出せます。  
内標のピーク位置がサンプルと途中で前後したりするときは注意。

````python
idx_sample = (16,135)   # 簡単のため透過実験のレシーバー側のやつだけ使います。(model2を使う、model1はドナー側用らしい。多分)

drug = "Peak-1"
IS = "Peak-2"

df_sample = df.loc[idx_sample[0]:idx_sample[1],:]
ratio_sample = df_sample.loc[:,drug]/df_sample.loc[:,IS]
ratio_sample.name = "ratio"
df_sample = pd.concat((df_sample, ratio_sample), axis=1)
df_sample = df_sample.reset_index(drop=True)
df_sample
````

````python
pred_sample = (df_sample.loc[:,"ratio"] - model2.intercept_[0]) / model2.coef_[0,0]   #x = (y-b)/aより
pred_sample.name = "prediction"
df_summary = pd.concat((df_sample, pred_sample), axis=1)
df_summary
````

ではいよいよ実験サンプルごとに薬物濃度と累積濃度をプロットしていきます。  
まず、サンプルごとのインデックスを作ります。

````python
base_names = ["A","B", "C", "D", "E", "F", "G", "H"]
timepoints = [0,1,2,3,4,5,6,7,8,22,23,24,25,26,27]

names = []
for base_name in base_names:
    temp_names = []
    for t in timepoints:
        temp_names.append(base_name+str(t))   #strで数値をstring型へ変換
    names.append(temp_names)
print(names)
````

では、これを使ってもう少しデータ整形。  
また、データフレームにtimepointsも追加してしまえば、後々使いやすいです。  
参考までに、サンプルAを例に挙げてから全体の処理を書きます。

サンプルAの例  
以下のように切り出せます。

````python
a=df_summary.set_index("ｸﾛﾏﾄ名").loc[names[0]]
time = pd.Series(timepoints, name="time", index=a.index)   # Series, DataFrame, Panelの順に、1次元、2次元、3次元の型となっています。扱い方は基本同じ。
a=pd.concat((a,time), axis=1)
a
````

もうサンプルAだけのデータフレームなので、クロマト名とか要らないですね。  
timeをインデックスにしてしまいましょう。  

````python
a=a.set_index("time", drop=True)   # drop=Trueで、現在のインデックスを上書き（データフレームに戻さないということ）
a
````

はいできました。ついでにpredictionの累積もとって、データフレームにくっつけましょう。  
cumsumという便利な関数があります。

````python
cum = a.loc[:,"prediction"].cumsum()
cum.name = "cumsum"
a = pd.concat((a,cum), axis=1)
a
````

こんな感じで解析を進めることができます。ではループを回してすべてのサンプルに同様の処理を行います。  
また、すべて同じサイズのデータフレームなので、3次元のパネルにしたほうが扱いやすいですね。  
以下では加えて以下の事も行います。  
ほぼ解析シートでやってることと同じだと思います。

1. time 0 では薬物濃度0のはずなので、補正
1. 体積も考慮して薬物濃度から薬物量を計算
1. 単位面積当たりの累積透過量を計算
1. 単位面積当たり、単位時間当たりの累積透過量を計算（Fluxに相当）

他にやりたいことがあったら、自分で追加できるようにしましょう。  
サンプルから薬物の抽出をしている場合には抽出率で薬物量を補正する必要があるけど、  
計算が面倒なので今回既に計算されたものとして使います。  
自分で使うときはその部分もプログラムに書いておくとよいでしょう。

````python
from collections import OrderedDict


area = 1.77
V_sample = 0.5   # 500uLサンプルしたのでmL換算した
V_receiver = [5.9, 5.6, 5.9, 5.6, 6.2, 5.8, 5.6,6.0]   # 今回は手打ちしたが、csvとかに書き込んでおいて、読み込んだ方が楽かも（人に渡すのも）
ratio_extract = 0.907479267   # これは計算されたものとして使っている。

results = OrderedDict()

for i,name in enumerate(names):
    V = V_receiver[i]
    # データ切り出し
    df_sample_each = df_summary.set_index("ｸﾛﾏﾄ名").loc[name]
    
    # timeを追加してインデックスに指定
    time = pd.Series(timepoints, name="time", index=df_sample_each.index)   #timeは使いまわせるのでループの外でやった方がいいけど、今回は気にしない。
    df_sample_each = pd.concat((df_sample_each,time), axis=1)
    df_sample_each = df_sample_each.set_index("time", drop=True)
    
    # 濃度の補正
    corrected = (df_sample_each.loc[:, "prediction"] - df_sample_each.loc[0,"prediction"])/ratio_extract   #補正, カラムの追加はこんな書き方も可能
    corrected.name = "prediction_correct"
    df_sample_each = pd.concat((df_sample_each, corrected), axis=1)
    
    # 補正後濃度からサンプル中およびレシーバー薬物量を算出
    amount_sample = df_sample_each.loc[:,"prediction_correct"] * V_sample
    amount_receiver = df_sample_each.loc[:,"prediction_correct"] * V
    amount_sample.name = "amount_sample"
    amount_receiver.name = "amount_receiver"
    
    # サンプル中およびレシーバー薬物量から累積透過量を算出（単位面積あたりも算出）
    cum_sample = amount_sample.cumsum()
    cum_permeated = amount_receiver + cum_sample
    cum_permeated_per_area = cum_permeated / area
    cum_sample.name = "cumsum_sample"
    cum_permeated.name = "cumsum_permeated"
    cum_permeated_per_area.name = "cumsum_permeater_per_area"
    
    # 累積透過量から単位時間当たりの透過量を算出（hourをsecに変換するため3600倍してる）
    flux = cum_permeated_per_area.diff()/(time.diff().values * 3600)   # diffで前後の差分を計算してくれる。valuesにしているのはindexが同じでないので、index情報がないnumpyにするため。
    flux.name = "flux"
    
    #　計算結果を1つのデータフレームに合体
    results[base_names[i]] = pd.concat((df_sample_each, amount_sample, amount_receiver, cum_sample, cum_permeated, cum_permeated_per_area), axis=1)
pnl = pd.Panel(results)
````

````python
pnl.to_excel(os.path.join(dir_name, "results.xlsx"))   #Panelはエクセルで保存するとitemsごとにシートを分けて保存してくれる。
````
