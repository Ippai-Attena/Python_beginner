## numpyとどう違うの？

numpyはnumeric pythonの略。  
数値計算を行うためのライブラリなので、数値の入った配列、ベクトルや行列を扱うことを前提としています。  
計算するだけなら正直numpyで十分ですが、列ごとに名前がついていたり、行ごとにインデックスが付いていた方が便利なことがあるでしょう。特にデータ解析の現場では、そのような機会に出くわすことがほとんどです。

そんなとき、行と列にそれぞれインデックスを付けてnumpyの配列を扱いやすくしたものがあるととても便利です。  
それがPandas。
「行と列それぞれに名前がついてるだけ以外はnumpyと一緒じゃん！」とか思ったらその通りです。  
pandasはnumpyを使って作られています。  
なので、扱い方は非常に似てるし、データフレームから数値の配列だけを取り出してnumpyのarrayにしてしまうことも容易です。  

じゃあ早速便利さを体感していきましょう。

説明のため、機械学習ライブラリscikit-learnにデフォルトで用意されているデータを使います。  
__データ読み込み__(ここは深く考えなくていいです)

````python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
%matplotlib inline

iris = load_iris()
````

````python
# numpy arrayでデータが入っている
iris.data
````

````python
# irisのfeature_namesにはデータの各列が示すものが入っている
iris.feature_names
````

さて、petal lengthだけの値を取り出したいなーって時に、どう取り出しますかね
```
iris.data[:,2]   # [行, 列]で指定。コロン:は「全て」という意味
```
で取り出せるけど、いつもこの番号覚えてなきゃいけないなんて辛すぎないですか？  
たった4列しかないならいいけど、数百、数千となったらもう無理です。  
名前で指定できるようにpandasでデータフレーム化しましょう。

````python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df
````

はい。まず劇的に見やすいですね！  
ではpetal length (cm)だけ取り出してみましょう。

````python
df.loc[:,"petal length (cm)"]
````

簡単に取れだせました。  
[ここから:ここまで]のような指定の仕方もできます。

````python
df.loc[:,"sepal width (cm)":"petal width (cm)"]
````

途中がなくてもリストを渡せば抽出可能

````python
select = ['sepal length (cm)',
          #'sepal width (cm)',　# これだけ要らないときとか
          'petal length (cm)',
          'petal width (cm)']
df.loc[:,select]
````

もちろん行方向でもできます

````python
df.loc[0:10, select]
````

## 注意点

locは、__名前で指定する__ことを前提としています。  
行方向のindexは数字ですが、これも各行の「名前」としてスライスします。  
Pythonのスライスは[以上:未満]でしたよね？（arangeを思い出してください）  
名前に以上も未満もありませんので、locでは指定した範囲をすべて含みます。  

## 番号で指定したいこともあるんですけど？

ilocがあります。

````python
df.iloc[0:10, 0:3]
````

これは[以上：未満]で指定することになるので注意しましょう。  
また、ilocは番号で指定するメソッドなので、名前での指定はできません。  
もし、行は番号で、列は名前で指定したいなら順番にやるしかないですね。

````python
df.iloc[0:10,:].loc[:,select]
````

## Pandasは実は統計解析をするためのライブラリ
（もっというと時系列解析だけど。株価の変動とか得意。）

なので、統計解析用のメソッドも結構充実しています。  
プロットも簡単にできるし、使いこなせれば非常に便利！  
実演していきます。

````python
print("行方向へ平均")
print(df.mean(axis=0))
print("行方向へ不偏分散")   # 不偏分散が何かは各々勉強してください。
print(df.var(axis=0, ddof=1))   # ddof: degrees of freedom（自由度）
print("行方向へ不偏標準偏差")
print(df.std(axis=0, ddof=1))   # ddof: degrees of freedom（自由度）
````

axis=1にすれば列方向にもできます。  
他にも移動平均、移動分散なんかもできます。  
実は全組み合わせの相関も一撃で計算できます（numpyでもできる）。

````python
df.corr(method="pearson")
````

````python
df.corr(method="spearman")
````

相関には__ピアソンの相関係数__と__スピアマンの相関係数__があるので、使い分けることができます。  
初めに習う相関はピアソンの方です。  
知らなかった場合はスピアマンの相関係数も勉強しておきましょう。
