## パッケージの呼び出し方
例えばだけど

````python
import numpy as np
````

とすれば numpy を npという名前で呼び出すという意味になります。npの部分はプログラム中で使う別名で、自分で好きな名前を付けたら良いです（numpyって毎回打つのは辛いよねってことだけです）。ただ、大体の場合は業界標準があるので真似したほうが絶対に良いです。


他にも一部の機能だけ呼び題したい場合は

````python
from numpy import array
````

のように、numpyが持っているarrayという機能を呼び出すということになります。

これも最後にas hogehogeとすれば別名を付けられます。

というわけで、pythonに元々提供されていない機能を使うときにパッケージを呼び出すことが分かったと思うので、細かい使い方は置いといてグラフをプロットしてみます。

# matplotlibでプロットしてみる

````python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline 
#これはグラフをここにデフォルト表示するためのコマンドです。おまじないだと思って書いといてください。わかんなくていいです。
````

````python
x = np.linspace(0,4*np.pi,1000)
y = np.sin(x)
````

````python
plt.plot(x,y)
````
