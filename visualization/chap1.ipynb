# matplotlibでプロットしてみる

````python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline 
#これはグラフをここにデフォルト表示するためのコマンドです。おまじないだと思って書いといてください。わかんなくていいです。
````

````python
x = np.linspace(0,4*np.pi,1000)   #0-4piまでを1000分割
y = np.sin(x)   #sin(x)を計算。（1000個の値が出てくる）
plt.plot(x,y)
````

````python
x = np.linspace(0,4*np.pi,20)   #0-4piまでを20分割
y = np.sin(x)   #sin(x)を計算。（20個の値が出てくる）
plt.plot(x,y)
````

````python
x = np.linspace(0,4*np.pi,1000)   #0-4piまでを1000分割
for i in range(10):   # 0-9まで順番に値を取り出しでiに代入
    y = np.sin(x-(np.pi/10)*i)
    plt.plot(x,y)
````

のように使用します。

__Q. np.sinとかplt.plotとかの「.」ってどんな意味？__  
「.」は所属を表しています。np.sinはnpの中にあるsin関数って意味です。  
sinは元々pythonにありません（あるにはあるけど、mathを別途呼び出すのでやることは同じ）。  
なので、どこからどんな機能を呼び出すのか明確にしてコードを書く必要があります。

__Q. なんでそんな面倒な仕様になってるの?__  
プログラミング言語は「関数型言語」と「オブジェクト指向型言語」に大別されます。  
関数型の代表例としてはC言語が、オブジェクト指向型の代表としてPythonがあげられます。  
関数型言語では、何か入力を与えたら、なんか出力を吐き出す「処理（関数）」をコーディングしていきます。  

一方で「オブジェクト指向型」は、データと関数をまとめた「オブジェクト」というものを考えます。
必ずしもデータと関数がセットになったものじゃなくてもいいんですが、とにかく何かモノ（実体）があって、そいつを弄繰り回すイメージです。  

  
  
  
たとえ話をすると...  
関数型でいう関数とは入力に対して処理を行う処理部分のことで、日常生活で例えるなら「ハサミ」とか「ノリ」とか「ホチキス」といったものに相当します。切ったり、貼ったり、まとめたりするわけです。当然紙以外にこんなことはできないので、石や鉄を渡そうとすると「無理っす（ERROR）」となるわけです。

オブジェクト指向型では、まず最初に「モノ（紙）」を考えます。この紙にどんな処理をするかを記述します。なので、まず紙があって、それを「投げる」といった処理は「そんな事できる方法（関数）がないよ！（ERROR）」となるわけです。

で、「なんでそんな面倒な仕様になってるの?」に戻ると、先にモノがあればできることが限定されるので、難しいことを考える必要がないわけです。  
「ハサミ、ノリ、ホチキスで何ができる？」って言われるより、「紙に対してできることは？」って言われた方ができることのイメージはしやすくないですか？？？だから、「.」を使って何に対する処理なのか所属を明らかにするんです。


...何となくはわかったかな？  
「オブジェクト指向ってなんだよ！」ってなっている人は結構見かけるので、わからなくても問題はないです。  
一言でまとめるなら「モノが先か、道具が先か」って感じでしょうか。

````python
plt.rc('font', size=10)  #説明のために、一旦元のサイズに戻してるだけです。

fs = 14
plt.rc('xtick', labelsize=fs)   # matplotlibのデフォルト設定をいじる（x軸）
plt.rc('ytick', labelsize=fs)   # matplotlibのデフォルト設定をいじる（y軸）


plt.scatter(x,y)
fs = 20  # font sizeを指定した
plt.title("sine curve", fontsize=fs)
plt.xlabel("x axis", fontsize=fs)
plt.ylabel("y axis", fontsize=fs)
````

### 表示するx軸とy軸の範囲を指定したい

````python
plt.scatter(x,y)
plt.xlim(1,8)   #x軸の1-8の範囲だけ表示
plt.ylim(0,1)   #y軸の0-1の範囲だけ表示
````

### 正方形にしたい

````python
plt.scatter(x,y)
ax = plt.gca()  #現在アクティブなグラフのオブジェクト（実体）をとってくる。get current axisの略
xlim = ax.get_xlim()   #描画されているx軸の最小値と最大値
ylim = ax.get_ylim()   ##描画されているy軸の最小値と最大値

xrange = xlim[1]-xlim[0]   #xlim[0]はxlimに入っている0番目の値をとってこいって意味。
yrange = ylim[1]-ylim[0]


ax.set_aspect(xrange/yrange)   # xの範囲/yの範囲って渡す
````

### ここまでのまとめ
グラフを正方形にして見栄え良くしてみます

````python
font = {"family": "serif"}   #ちゃっかりフォントの種類も変えた
plt.rc("font", **font)
plt.rc('xtick', labelsize=16)   # matplotlibのデフォルト設定をいじる（x軸）
plt.rc('ytick', labelsize=16)   # matplotlibのデフォルト設定をいじる（y軸）

plt.figure(figsize=(7,7))
plt.plot(x,y, color="k", lw=3, ls="--")   #ls: line styleの略"--"で破線になる。kは黒という意味。*kuroとは関係ないです。
plt.scatter(x,y, s=100, color="blue")
plt.legend(["sine curve"], fontsize=14)

fs = 20  # font sizeを指定した
plt.title("sine curve", fontsize=fs)
plt.xlabel("$x$ axis", fontsize=fs)
plt.ylabel("$y$ axis", fontsize=fs)



ax = plt.gca()  #現在アクティブなグラフのオブジェクト（実体）をとってくる。get current axisの略
xlim = ax.get_xlim()   #描画されているx軸の最小値と最大値
ylim = ax.get_ylim()   ##描画されているy軸の最小値と最大値
xrange = xlim[1]-xlim[0]   #xlim[0]はxlimに入っている0番目の値をとってこいって意味。
yrange = ylim[1]-ylim[0]
ax.set_aspect(xrange/yrange)   # xの範囲/yの範囲って渡す
````
